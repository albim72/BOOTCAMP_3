```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,2.0,0.01)
s = 1 + np.sin(2*np.pi*t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel='czas [s]',ylabel='napięcie prądu [mv]', title="pomiar napięcia w czasie")
ax.grid()
fig.savefig("mwykres.png")
plt.show()
```

```python
#daklracja danych
x1 = np.linspace(0.0,5)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)

x2 = np.linspace(0.0,2)
y2 = np.cos(2*np.pi*x2)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1)
fig.suptitle("dwa wykresu:; drgania tłumione i nietłumione.")

ax1.plot(x1,y1,'o-')
ax1.set_ylabel('drgania tlumione')

ax2.plot(x2,y2,'.-')
ax2.set_ylabel('drgania nietlumione')
ax2.set_xlabel('czas [s]')

plt.show()
```
