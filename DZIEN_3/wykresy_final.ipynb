```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
import matplotlib as mpl
```

# wykres słupkowy

```python
fig,ax = plt.subplots()
marki = ['audi','toyota','mercedes','honda','kia','jeep','seat']
sprzedaz = [12.7,28.5,4.6,16.8,21.3,6.2,14]

bar_labels = ['audi','toyota','mercedes','honda','kia','jeep','seat']
bar_colors = ['tab:red','tab:green','tab:red','#A3B456','tab:orange','tab:blue','#FFFF00']

ax.bar(marki,sprzedaz,label=bar_labels,color=bar_colors)
ax.set_xlabel('marki samochodów')
ax.set_ylabel('sprzedaż samochodów')

ax.legend(title='marki')
plt.show()
```

# wykres kołowy

```python
labels = 'Warszawa','Kraków','Lublin','Katowice','Gdańsk'
size = [56,34,37,24,41]
explode = (0,0.1,0,0,0.2)
```

# histogram

```python
np.random.seed(19680801)
mu = 100 #średnia rozkładu
sigma = 15
x = mu+sigma*np.random.randn(437)
```

```python
num_bins = 50
fig,ax = plt.subplots()
n,bins,patches = ax.hist(x,num_bins,density=True)

#wzór na rozkład normalny (funkcja gęstości prawdopdobieństwa)

y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*(1/sigma*(bins-mu))**2))

ax.plot(bins,y,'--')
ax.set_xlabel('wartości')
ax.set_ylabel('gęstość prawdopodobieństwa')
ax.set_title('Histogram wartość Q: $\mu=100$, $\sigma=15$')
fig.tight_layout()
plt.show()
```

# wykres typu: Cykle

```python
cases = [
    None,
    8,
    (30,8),
    [16,24,30],
    [0,-1],
    slice(200,200,3),
    0.1,
    0.3,
    (0.0,0.1),
    (0.45,0.1)
]

colors = [
    '#1f77b2',
    '#ff7f0c',
    '#2ca02c',
    '#d62728',
    '#9467bd',
    '#8c564b',
    '#e377c2',
    '#7f7f7f',
    '#bcbd22',
    '#17becf'
]
```

```python
mpl.rcParams['axes.prop_cycle'] = cycler(markevery=cases,color=colors)
x = np.linspace(0,2*np.pi)
offsets = np.linspace(0,2*np.pi,11,endpoint=False)
yy = np.transpose([np.sin(x+phi) for phi in offsets])

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.6,0.75])

for i in range(len(cases)):
    ax.plot(yy[:,i],marker='o',label=str(cases[i]))
    ax.legend(bbox_to_anchor=(1.05,1),loc='upper left', borderaxespad=0.)

plt.title('wyświetlanie grupy wykresów cyklicznych...')
plt.show()
```
